```python
import pandas as pd
from bs4 import BeautifulSoup
from requests import get
import re
```

```python
url = "https://www.imdb.com/search/title/?num_votes=100000,&sort=user_rating,desc&title_type=tv_series&start=1"
```

```python
html_format=get(url)
```

```python
bs4_obj = BeautifulSoup(html_format.text, "html.parser")

bs4_obj
```

```python
master_container = bs4_obj.find_all('div', class_ = "lister-item mode-advanced")

len(master_container)
```

```python
movie_name = []
for card in master_container:
    content = card.find('div', class_='lister-item-content')
    h3 = content.find('h3', class_='lister-item-header')
    movie_name.append( h3.find('a').text)
    
    
movie_name
```

```python
movie_name = []
release_year = []
rating = []
votes = []
star_cast = []

for x in range(1,100, 50):
    url = "https://www.imdb.com/search/title/?title_type=tv_series&num_votes=100000,&sort=user_rating,desc&ref_=adv_prv"+str(x)
    html_format=get(url)

    bs4_obj = BeautifulSoup(html_format.text, "html.parser")

    #bs4_obj

    master_container = bs4_obj.find_all('div', class_ = "lister-item mode-advanced")

    #len(master_container)
    for card in master_container:
        content = card.find('div', class_='lister-item-content')

        # For Realease Year
        release_year.append(card.find('span', class_='lister-item-year').text)


        # For Movie Name ratings-imdb-rating
        h3 = content.find('h3', class_='lister-item-header')
        movie_name.append( h3.find('a').text)

        # For Rating
        rating_div = content.find('div', class_='ratings-imdb-rating')
        rating.append(rating_div.find('strong').text)

        #For votes
        vote_con=content.find('p',class_='sort-num_votes-visible')
        votes.append(vote_con.find('span',{'name':'nv'}).text)
        
        #for movie cast
        movie_cast=card.find('p',class_='').text
        movie_cast1=re.sub(r"\n","",movie_cast)
        star_cast1=movie_cast1.partition('Stars:')[2]
        star_cast.append(star_cast1)
    
#release_year
#movie_name
#rating
votes

df1 = pd.DataFrame({"Movie Name" : movie_name,
                    "Release year":release_year,
                  "Rating" : rating,
                   "votes":votes,
                   "star cast":star_cast})
df1
```
