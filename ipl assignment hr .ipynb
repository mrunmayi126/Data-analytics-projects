Find the top 10 batsman based on batting average (which is total runs by total matches played).

```python
avg_bat=new_df.groupby(['Player_Name']).agg({'Batsman_Scored':'sum',
                                   'Match_Id':'count'}).reset_index().sort_values(by='Batsman_Scored',ascending= False)
avg_bat.head(10)
```

```python
final_players_cent=players_cent.sort_values(by='Batsman_Scored',ascending= False).head(10)
final_players_cent
```
